Apart from the provided feature variables, we can extract Mr, Mrs, Miss and Master from the Names to derive a new categorical variable. This variable should be useful in predicting Age. It can replace Sex.

```python
import re

train_set_transformed = train_set.copy()
train_set_transformed["Title"] = pd.Series(np.nan, index=train_set_transformed.index)
title_idx = train_set_transformed.columns.get_loc("Title")

for index, row in train_set_transformed.iterrows():
    m = re.search('Mrs|Mr|Miss|Master', row["Name"])
    if m:
        train_set_transformed.iloc[index, title_idx] = m.group(0)
```

```python
import random

random.seed(702)
train_set_transformed.iloc[random.sample(list(range(len(train_set_transformed))), 25)][["Name", "Title"]]
```

```python
# Check how many NaN there are (24).

train_set_transformed[["Title"]].info()
```

```python
train_set_transformed.loc[pd.isna(train_set_transformed["Title"])]["Name"]
```

```python
# Set all the Rev to Title=Mr

for index, row in train_set_transformed.loc[pd.isna(train_set_transformed["Title"])].iterrows():
    m = re.search('Rev', row["Name"])
    if m:
        train_set_transformed.loc[index, title_idx] = 'Mr'
```

```python
train_set_transformed.loc[pd.isna(train_set_transformed["Title"])][["Name", "Title"]]
```
